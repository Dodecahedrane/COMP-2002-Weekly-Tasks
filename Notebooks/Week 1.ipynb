```python
#Bubble Sort

toSort = [1,4,6,22,4,6,13,41,7,9,23,57,3]

for i in range(len(toSort)):
    for j in range(len(toSort) - 1):
            if toSort[j] > toSort[j+1]:
                toSort[j], toSort[j+1] = toSort[j+1], toSort[j]

print(toSort)
```

```python
#Print the first N digits of the Fibonacci sequence

fibonacci = [1,1]

n = 10

for i in range(n-2):
    fibonacci.append(fibonacci[i] + fibonacci[i+1])
    
print(fibonacci)
```

```python
#Generate a plot of the function y = sin(x) between the values -π and π
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, num=100)
y = np.sin(x)

plt.plot(x, y, color="green")
#plt.scatter(x, y, color="red")
```
